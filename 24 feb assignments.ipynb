## Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.



```python
import pandas as pd

df = pd.DataFrame({'A': [10, 20, 30], 'B': [40, 50, 60], 'C': [70, 80, 90]})
def reindex_dataframe(df):
    df = df.reset_index(drop=True)  # Reset existing index
    df.index = pd.RangeIndex(start=1, step=2, stop=len(df) * 2)  # Create new index
    return df

new_df = reindex_dataframe(df)
print(new_df)

```

## Q5. How are DataFrame.size() and DataFrame.shape() different?
1. DataFrame.size:

* DataFrame.size returns the total number of elements in the DataFrame, which is calculated as the number of rows multiplied by the number of columns.
* It represents the total number of values in the DataFrame, including missing or NaN values.
* The result is an integer value.
* The attribute is accessed without parentheses, i.e., DataFrame.size, not DataFrame.size().

2. DataFrame.shape:

* DataFrame.shape returns a tuple representing the dimensions of the DataFrame.
* The tuple contains two elements: the number of rows and the number of columns, respectively.
* It provides an overview of the DataFrame's structure.
* The result is a tuple of integers.
* The attribute is accessed with parentheses, i.e., DataFrame.shape().

```python
## Example
import pandas as pd

df = pd.DataFrame({'A': [10, 20, 30], 'B': [40, 50, 60], 'C': [70, 80, 90]})

size = df.size
print("DataFrame.size:", size)

shape = df.shape
print("DataFrame.shape:", shape)
```

## Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.

## The username is the part of the email address that appears before the '@' symbol. For example, if the email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your function should extract the username from each email address and store it in the new 'Username' column.

```python
df = pd.DataFrame({'Email': [
    "john.doe@example.com",
    "jane.smith@example.com",
    "david.johnson@example.com",
    "emily.wilson@example.com",
    "michael.brown@example.com",
    "sarah.davis@example.com",
    "alexander.thomas@example.com",
    "olivia.martin@example.com",
    "william.jackson@example.com",
    "sophia.anderson@example.com"
]})

def create_col(df):
    a = []
    for i in df['Email']:
        w = list(i.split("@"))
        a.append(w[0])
        
    df['Username'] = a
    return df

create_col(df)
```

## Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows.

For example, if df contains the following values:

   A   B   C

0  3   5   1

1  8   2   7

2  6   9   4

3  2   3   5

4  9   1   2

Your function should select the following rows:   A   B   C

1  8   2   7

4  9   1   2

The function should return a new DataFrame that contains only the selected rows.

```python
import pandas as pd

data = {'A': [3, 8, 6, 2, 9],
        'B': [5, 2, 9, 3, 1],
        'C': [1, 7, 4, 5, 2]}

df = pd.DataFrame(data)

def select_rows(df):
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    return selected_rows

select_rows(df)
```

## Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
import pandas as pd

# Example DataFrame
data = {'Date': pd.date_range(start='2023-01-01', periods=10),
        'Sales': [10, 20, 15, 25, 30, 35, 20, 15, 10, 5]}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print()

# Function to calculate moving average
def calculate_moving_average(df):
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    return df

# Call the function to calculate moving average
new_df = calculate_moving_average(df)
print("DataFrame with Moving Average:")
print(new_df)
```

## Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.

For example, if df contains the following values:

         Date

0  2023-01-01

1  2023-01-02

2  2023-01-03

3  2023-01-04

4  2023-01-05

Your function should create the following DataFrame:


         Date    Weekday

0  2023-01-01    Sunday

1  2023-01-02     Monday

2  2023-01-03    Tuesday

3  2023-01-04    Wednesday

4  2023-01-05    Thursday

The function should return the modified DataFrame.

```python
import pandas as pd

# Example DataFrame
data = {'Date': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'])}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print()

# Function to add weekday column
def add_weekday_column(df):
    df['Weekday'] = df['Date'].dt.day_name()
    return df

# Call the function to add weekday column
new_df = add_weekday_column(df)
print("DataFrame with Weekday column:")
print(new_df)
```

## Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.



```python
import pandas as pd

# Example DataFrame
data = {'Date': pd.to_datetime(['2023-01-01', '2023-01-15', '2023-01-25', '2023-02-05'])}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print()

# Function to select rows between dates
def select_rows_between_dates(df, start_date, end_date):
    mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    selected_rows = df[mask]
    return selected_rows

# Call the function to select rows between dates
selected_rows = select_rows_between_dates(df, '2023-01-01', '2023-01-31')
print("Selected Rows:")
print(selected_rows)
```
