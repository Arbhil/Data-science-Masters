## Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

In pandas, a DataFrame is a two-dimensional tabular data structure that represents data in a rows-and-columns format. It is similar to a table in a relational database or a spreadsheet in Excel. A DataFrame can store and manipulate heterogeneous data types, such as integers, floats, strings, etc., and is a versatile data structure for data analysis and manipulation.

On the other hand, a Series in pandas is a one-dimensional labeled array capable of holding any data type. It can be thought of as a single column of a DataFrame. A Series provides an index to access each element, similar to a dictionary, and can be created from various data sources like lists, arrays, or dictionaries.

Here's an example to illustrate the difference between a DataFrame and a Series:

```python
import pandas as pd

# Creating a DataFrame
data = {
    'Name': ['John', 'Alice', 'Bob'],
    'Age': [25, 30, 35],
    'Country': ['USA', 'Canada', 'UK']
}
df = pd.DataFrame(data)

# Creating a Series
series = pd.Series([4, 8, 15, 16, 23, 42])

print("DataFrame:")
print(df)
print("\nSeries:")
print(series)
```

## Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd

# Creating the Series
names = pd.Series(['John', 'Alice', 'Bob'])
ages = pd.Series([25, 30, 35])
countries = pd.Series(['USA', 'Canada', 'UK'])

# Creating the DataFrame
data = {
    'Name': names,
    'Age': ages,
    'Country': countries
}
df = pd.DataFrame(data)

# Printing the DataFrame
print(df)
```
